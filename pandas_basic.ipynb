```python
data = {
    "country":["brazil","Rassia","India","china","south africa"],
    "population":[200.4,252.5,2525.4,888.5,96.5],
    "gdp":[1.86,1.3,2.6,11.4,0.29],
    "area":[85167,17058,8567,8567,8589],
}

cdf=pd.DataFrame(data)
cdf
```

```python
print(cdf['gdp'].mean())
```

```python
cdf.plot(kind='bar', x='country', y='population',figsize=(6,4))
```

```python
cdf.plot(kind='bar', x='country', y='population',logy=True)
```

```python
cdf.plot(kind='bar', x='country', y='population',logy=True,
         color='green',
         alpha=0.5,
         hatch='//',
         edgecolor='white')
```
